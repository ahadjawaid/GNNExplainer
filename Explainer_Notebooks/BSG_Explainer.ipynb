```python
def createMaskMeanDict(mask):
    maskMean = {}
    for key in mask:
        if list(mask[key]):
            maskMean[key] =  torch.mean(mask[key])
    return maskMean

def getMaxKey(dictionary):
    maxKey = None
    maxValue = 0
    for key in dictionary:
        if dictionary[key] > maxValue:
            maxValue = dictionary[key]
            maxKey = key
    
    return maxKey, maxValue
```
